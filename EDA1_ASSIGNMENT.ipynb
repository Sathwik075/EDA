```python
lb_med = df_cleaned['LB'].median()
ac_med = df_cleaned['AC'].median()
fm_med = df_cleaned['FM'].median()
uc_med = df_cleaned['UC'].median()
dl_med = df_cleaned['DL'].median()
ds_med = df_cleaned['DS'].median()
dp_med = df_cleaned['DP'].median()
astv_med = df_cleaned['ASTV'].median()
mstv_med = df_cleaned['MSTV'].median()
altv_med = df_cleaned['ALTV'].median()
mltv_med = df_cleaned['MLTV'].median()
width_med = df_cleaned['Width'].median()
tendancy_med = df_cleaned['Tendency'].median()
nsp_med = df_cleaned['NSP'].median()
```

```python
df_cleaned['AC']=df_cleaned['AC'].fillna(ac_med)
df_cleaned['LB']=df_cleaned['LB'].fillna(lb_med)
df_cleaned['FM']=df_cleaned['FM'].fillna(fm_med)
df_cleaned['UC']=df_cleaned['UC'].fillna(uc_med)
df_cleaned['DL']=df_cleaned['DL'].fillna(dl_med)
df_cleaned['DS']=df_cleaned['DS'].fillna(ds_med)
df_cleaned['DP']=df_cleaned['DP'].fillna(dp_med)
df_cleaned['ASTV']=df_cleaned['ASTV'].fillna(astv_med)
df_cleaned['MSTV']=df_cleaned['MSTV'].fillna(mstv_med)
df_cleaned['ALTV']=df_cleaned['ALTV'].fillna(altv_med)
df_cleaned['MLTV']=df_cleaned['MLTV'].fillna(mltv_med)
df_cleaned['Width']=df_cleaned['Width'].fillna(width_med)
df_cleaned['Tendency']=df_cleaned['Tendency'].fillna(tendancy_med)
df_cleaned['NSP']=df_cleaned['NSP'].fillna(nsp_med)
```

Here all the null values are filled with their meadian because all the columns contains the outilers if outilers are contain we should go for the median if no mean.

```python
df_cleaned.isnull().sum()
```

# Treating with the outilers

```python
box_ac = plt.boxplot(df_cleaned['AC'])
extermes = [item.get_ydata()[1]for item in box_ac['whiskers']]
UE = extermes[0]
LE = extermes[1]
print(LE,UE)
```

```python
df_cleaned.loc[df_cleaned.AC>UE,'AC']=UE
df_cleaned.loc[df_cleaned.AC<LE,'AC']=LE
```

```python
plt.boxplot(df_cleaned.AC)
```
